```python
# Standard parsing to decode FROM JSON creates dictionaries or lists
# supply `object_pairs_hook` or `object_hook` to the call to
# json.loads()

# ordered dictionary specified:
s = '{"name" : "ACME", "shares" : 50, "price" : 490.1}'
from collections import OrderedDict
data = json.loads(s, object_pairs_hook=OrderedDict)
```

```python
data
```

```python
# What if a JSON dictionary needs to become a Python object?

class JSONObject:
    def __init__(self, d):
        self.__dict__ = d

data = json.loads(s, object_hook=JSONObject)
data.name
```

```python
# The above didn't work...

# Running into trouble when trying to serialize Python data instances
# to be ready for JSONifying:

# suppose: want to serialize all instances of an class...
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# create an instance 'p'
p = Point(2, 3)

# try to JSONify it:
json.dumps(p)
```

```python
# A possible fix here could be to write a function that can
# accept instance but returns a dictionary. Dicts can be serialized

def series_ify(obj):
    d = { '__classname__' : type(obj).__name__ }
    d.update(vars(obj))  # <-- what is this vars up to here?
    return d

json.dumps(series_ify(p))
```

```python
# How about bring an instance back forom having been serialized?


# Specify a dictionary which includes the key and the value as
# the name of the class like so:

classes = {
    'Point' : Point
}

def unseries_ify(d):
    clsname = d.pop('__classname__', None)
    if clsname:
        cls = classes[clsname]
        obj = cls.__new__(cls)    # Doesn't call __init__
        for key, value in d.items():
            setattr(obj, key, value)
            return obj
    else:
        return d
```

```python
# Let's use these serialize and unserialize functions

p = Point(2, 3)
s = json.dumps(p, default=series_ify)
s
```

```python
a = json.loads(s, object_hook=unseries_ify)
a
```
